```python
data['Order Date'] = pd.to_datetime(data['Order Date'],dayfirst=True)
data['Ship Date'] = pd.to_datetime(data['Ship Date'],dayfirst=True)
```

```python
data.dtypes
```

```python
#method1 using apply for chaning datatype for multiple col
data[['Order Date','Ship Date']]=data[['Order Date','Ship Date']].apply(pd.to_datetime,dayfirst=False)
```

```python
data.dtypes
```

```python
#method2 using loop for chaning datatype for multiple col
for i in ['Order Date','Ship Date']:
  data[i]=pd.to_datetime(data[i])
```

```python
data.dtypes
```

```python
data['month']=data['Order Date'].dt.month
data['month']
```

```python
data['month']=data['Order Date'].dt.month_name()
data['month']
```

```python
data['month'] = pd.Categorical(
    data['month'],
    categories=[
        'January','February','March','April','May','June',
        'July','August','September','October','November','December'
    ],
    ordered=True
)
```

```python
monthly_sales=data.groupby('month')['Sales'].sum()
```

```python
import matplotlib.pyplot as plt
data.groupby('month')['Sales'].sum().plot(kind='line')
plt.title('sales by month')
plt.xlabel('month')
plt.ylabel('Sales')
plt.show()
```

```python
plt.xticks(range(len(monthly_sales.index)), monthly_sales.index, rotation=45)
```

```python
data.groupby('month')['Sales'].sum().plot(kind='line')
plt.xticks(rotation=45)
plt.title('Monthly Sales Trend')
plt.show()
```

```python
monthly_sales = data.groupby('month')['Sales'].sum().sort_index()

import matplotlib.pyplot as plt

monthly_sales.plot(kind='line')
plt.xticks(range(len(monthly_sales.index)), monthly_sales.index, rotation=45)
plt.title('Monthly Sales Trend')
plt.xlabel('month')
plt.ylabel('Sales')
plt.show()
```
